```python
# Question 4

# Most significant = calories and steps
# Least significant = calories and sedentary minutes
```

Most significant:
Null Hypothesis: There is no significant difference in the total number of steps and number of calories. 
Alternative Hypothesis: There is a significant difference in the total number of steps and number of calories. 

Least Significant:
Null Hypothesis: There is no significant difference in the sedentary minutes and number of calories.
Alternative Hypothesis: There is a significant difference in the number of sedentary minutes and number of calories. 

```python
import numpy as np
import scipy.stats as stats

middle_steps = daily_act["TotalSteps"].median()
first_sample = daily_act[daily_act["TotalSteps"] <= middle_steps]
second_sample = daily_act[daily_act["TotalSteps"] > middle_steps]
t_stat_cs, p_val_cs = stats.ttest_ind(first_sample["Calories"], second_sample["Calories"]) 
print(t_stat_cs, p_val_cs)
```

```python
middle_mins = daily_act["SedentaryMinutes"].median()
first_sample2 = daily_act[daily_act["SedentaryMinutes"] <= middle_mins]
second_sample2 = daily_act[daily_act["SedentaryMinutes"] > middle_mins]
t_stat_cs2, p_val_cs2 = stats.ttest_ind(first_sample2["Calories"], second_sample2["Calories"])
print(t_stat_cs2, p_val_cs2)
```
